```python
import pandas as pd
import numpy as np
```

```python
fama=pd.read_csv('fama-french factors.csv',index_col='date',parse_dates=True)
#fama=fama.set_index(pd.to_datetime(fama.index,format='%Y%m%d'))
weekly_return = fama.resample('W', label='left', loffset='1d').sum()
weekly_return=weekly_return*100
weekly_return['mom']=weekly_return['umd']-weekly_return['rf']
weekly_return['smb']=weekly_return['smb']-weekly_return['rf']
weekly_return['hml']=weekly_return['hml']-weekly_return['rf']
```

```python
ret=pd.read_csv('ew_returns.csv',index_col='date', parse_dates=True)
ret.head()
```

```python
data=pd.concat([ret.loc['2000-6-12':'2018-12-24',['excess_returns','excess_adj_returns']]*100,weekly_return.loc['2000-6-12':'2018-12-24','mktrf']],axis=1)
```

```python
def calc(data):
    n=len(data)
    m=max(data)
    mi=min(data)
    niu=sum(data)/n
    niu2=sum(data**2)/n
    niu3=sum(data**3)/n
    sigma = data.std()
    niu4=sum((data-niu)**4)/n
    skew =data.skew()
    kurt=data.kurt()
    ratio=(niu)/sigma
    return pd.Series({'max':m,'min':mi,'average': niu,'std': sigma,'kurtosis':kurt,'skewness':skew,'Sharpe ratio':ratio})
```

```python
metrics = pd.concat([calc(weekly_return[i]) for i in weekly_return.columns],axis=1)
metrics.columns=weekly_return.columns
metrics
```

```python
metrics.to_csv('metrics.csv')
```

```python
n=len(data)
wml=data['excess_returns']
RV=[]
for i in range(26,n):
    RV.append(sum((wml[i-26:i])**2)/26*4)
adj_return=wml[26:]*12/np.sqrt(12)/np.sqrt(RV)
```

```python
weight=12/np.sqrt(12)/np.array(np.sqrt(RV))
weight=pd.DataFrame(weight,index=adj_return.index)
plt.plot(weight,color='k')
plt.title('Adj Mom weight')
```

```python
from arch import arch_model
```

```python
am = arch_model(data['excess_returns'], vol='GARCH',p=1, o=1, q=1)
```

```python
forecast_vol=[]
#forecast_mean=[]
for i in range(len(data['excess_returns'])-26):
    res = am.fit(first_obs=i, last_obs=i+26, disp='off')
    temp = res.forecast().variance
    #temp_mean=res.forecast().mean
    fcast=temp.iloc[i+25:i+29].sum().values
    #fcast_mean=temp_mean.iloc[i+25:i+29].sum().values
    forecast_vol.append(np.sqrt(fcast[0]))
    #forecast_mean.append(fcast_mean[0])
```

```python
opt_w=12/np.sqrt(12)/forecast_vol
```

```python
opt_return=wml[26:]*opt_w
calc(opt_return)
```

```python
opt_w=pd.DataFrame(opt_w,index=opt_return.index)
opt_w.to_csv('opt_w.csv')
opt_w.head()
```
